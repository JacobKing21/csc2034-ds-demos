```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


dataset_red = pd.read_csv('winequality-red.csv', header=0)
dataset_white = pd.read_csv('winequality-white.csv', header=0)

print(f"Number of examples: {dataset_red.shape[0]}")
print(f"Number of features: {dataset_red.shape[1]}")
print(f"List of features:\n\t{dataset_red.columns}")
display(dataset_red.head())

print(f"Number of examples: {dataset_white.shape[0]}")
print(f"Number of features: {dataset_white.shape[1]}")
print(f"List of features:\n\t{dataset_white.columns}")
display(dataset_white.head())

plt.figure(figsize = (10, 6))
sns.countplot(x = "quality", data = dataset_red)
plt.title("Quality of red wine")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
plt.close()

plt.figure(figsize = (10, 6))
sns.countplot(x = "quality", data = dataset_white)
plt.title("Quality of white wine")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
plt.close()
```

From the graphs above you can see that white wine has 4898 examples compared to reds 1599. This could be a contributing factor as to why white wine has the highest score in quality, at 9. As described in winequality.names, there are much more normal wines scoring an average 5, 6 or 7 than there are excellent or poor wines.

```python
plt.figure(figsize = (25,6))
sns.countplot(x = "alcohol", data = dataset_red)
plt.title("Alcohol content in red wine")
plt.xticks(rotation=45) # Rotate the x labels so they don't overlap
plt.show()
plt.close()

plt.figure(figsize = (25,6))
sns.countplot(x = "alcohol", data = dataset_white)
plt.title("Alcohol content in white wine")
plt.xticks(rotation=45) # Rotate the x labels so they don't overlap
plt.show()
plt.close()
```
